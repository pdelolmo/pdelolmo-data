## APRENDIZAJE SUPERVISADO. LOS DATOS YA ESTAN ETIQUETADOS:

-REGRESION LINEAL: TENGO UN CONJUNTO DE PUNTOS Y TRENGO QUE ENCONTRAR LA RECTA QUE MAS APROXIME A LOS PUNTOS. LAS BETAS SON LOS COEFICIENTES. ESTOS COEFICIENTES MIDEN: SI YO AUMENTO X1 EN UN DETERMINADO VALOR CUANTO AUMENTARIA LA Y. HAY QUE CALCULAR LOS COEFICIENTES, LAS BETAS.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv')
df
```

```python
#vamos a hacer descripciones
df.info()
```

```python
df.isnull().sum()
```

```python
#EN REGRESION LINEAL, TENEMOS QUE DECIR CUAL ES LA VARIABLE QUE QUIERO PREDECIR Y CUALES VOY A TOMAR COMO DATO. A LAS QUE TOMO COMO DATO LLAMO X Y LA Y ES PARA ESOS DATOS
#LO QUE YO QUIERO PREDECIR, POR ESO SUPERVISADO. YO YA SE LO QUE TENGO. LAS QUE YO TOMO COMO DATO LAS LLAMAMOS X, (EN ESTE CASO SOLO UNA COLUMNA
#EXPRIENCE YEARS). X SIEMPRE SIEMRPE SIEMPRE LOS DATOS Y LA Y ES PARA ESOS DATOS, LO QUE YO QUIERO PREDECIR. AUNQUE YO TENGO YA LA SOLUCION PARA ESOS
#40 REGISTROS (YA SABEMOS SL SALARIO) POR ESO ES SUPERVISADO. DESPUES DE CALCULAR EL MODELO, PONDREMOS OTRO DATO DISTINTOS EN LOS AÑOS DE
#EXPERIENCIA Y VEREMOS CUANTO DINERO O CUAL SERI EL SALARIO, PREDECIR EE VALOR. QUEREMOS PREDECIR SALARIO (Y=SALARIO)
x=df[['Experience Years']] #2 corchetes porque solo nos interesa una lista de valores (para luego comparar)
y=df['Salary'] #aqui nos interesa continuar ocn el formato dataset
```

```python
x.shape ##RECOMENDABLE SIEMPRE!!!. NOS DIRA CUAL ES EL NUM DE FILAS Y DE COLUMNAS
```

##CUANTO TRATAMOS DE APRENDIZAJE NO SUPERVISADO, TODO EL CONJUNTO DE DATOS ES CON EL QUE USAMOS PARA CREAR EL MODELO. NO HACEMOS NINGUNA DIVISION. CUANDO TRATAMOS CON SUPERVISADO, SIN EMBARGO, DIVIDIMOS EL CNJUNTO DE DATOS EN 2 O 3 CONJUNTOS. HAY 2 QUE SIEMPRE VAN A ESTAR TRAIN SET Y TEST SET. TRAIN ES PARA GENERAR EL MODELO (DE ESTE CONJUNTO LE PASAREMOS LA SOLUCION Y LOS DATOS) Y EL TEST SET ES EL OTRO SUBCONJUNTO DE DATOS A LOS QUE LE PASAS LOS DATOS PERO NO LA SOLUCION. CALCULO CUANTO ES LA SOLUCION SEGUN EL MODELO Y COMPARO CON LA SOLUCION REAL QUE YO TENGO. PUEDE HABER UNA 3RA VARIANTE DE CONJUNTO DE VALIDACION (PERO HOY NO LO VEREMOS)


```python
#HAY QUE DIVIDIR LA MUESTRA EN 70/30 APROX
#PARA DIVIDIR HAY QUE USAR LA LUBRERIA TRAIN_TEST_SPLIT (dividimos entre entrenamiento y test)
#SIEMPRE PRIMERO TRAIN, DESPUE TEST, PRIMERO X, DESPUES Y !!!!!!
#PONEMOS LA FUNCION TRAINTESTSPLIT
#QUEREMOS DIVIDIR EL DATASET X Y LAS Y,
#TAMAÑO DE LA MUESTRA, 70/30% SIEMPRE O 80/20, SIEMPRE EN VALORES 0-1
#RANDOMSTATE- QUE NO COJA LAS COLUMNAS Y LAS FILAS EN ORDEN, SINO QUE LAS COJA DE MANERA ALEATORIA

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2529)
```

```python
##IMPORTANTE COMPROBAR UQE LOS VALORES SON LOS MISMOS
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
##ahora vamos a aplicar la regresion lineal,. DE SKLEARN (MODELO LINEAL, COGEMOS REGRESION LINEAL, APLICAMOS)
#IGUAL QUE CON ONEHOTENCODER, IGUAL CON KMEANS,
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model
```

```python
#AHORA QUEREMOS APLICAR LA REGRESION LINEAL A NUESTROS DATOS DE ENTRENAMIENTO (FIT PORQUE LO APLICAMOS) A LOS DATOS DE ENTRENAMIENTO
#COMO ESTAMOS EN EL ENTRENAMIENTO, TENEMOS QUE PASARLE LA SOLUCION (APRENDIZAJE SUPERVISADO) CAJITA AZUL XQ YA HEMOS APLICADO DATOS
model.fit(x_train,y_train)
```

```python
#YA PODEMOS VER LOS COEFICIENTES Y LA BETA0 QUE NO IBA CON NINGUNA VARIABLE (EL INTERCEPT), EL
#numero que va solo, no va con inguna X
model.intercept_
```

```python
#ESTA SERIA UNA LISTA DE LAS BETAS QUE MULTIPLICAN A LAS COLUMANS DE X. COMO SOLO TENEMOS UNA COLUMNA
#SOLO DA UN VALOR

model.coef_
```

```python
#estamos diciendo que la regresion son ambos coeficientes sumados * años de exp

26596+9405 * EXPERIENCE_YEARS
```

```python
#HEMOS COGIDO EL DATASET ETRENAMIENTO, HEMOS HECHO LA REGRESION LINEAL, HEMOS APLICADO LA REGRESION A LOS DATOS DE ENTRENAMIENTO
#TNEEMOS UN MODELO, PERO DEBEMOS SABER SI EL MODELO ES BUENO O NO Y DEBEMOS PREDECIR LOS VALORES DEL X TEST
#PARA PODER COMPARAR CON LA SOLUCION DEL XTEST
y_pred=model.predict(x_test)
y_pred
```

```python
#YA TENEMOS UNA SERIE DE VALORES, PERO COMO SABER SI LA REGRESION ES BUENA O NO
#SOBRE TODO 2 METRICAS: ERROR MEDIO (VALOR QUE ME DIO LA PREDICCION, VALOR QUE TENIA YO Y LO RESTAMOS, VALOR ABSOLUTO)
#ERROR CUADRATICO MEDIO: COMO HACER VARIANZA, LA REST PARA ELEVADA AL CUADRADO Y HACIENDO LA SUMA
#PARA ELLO HAY QUE IMPORTAR VARIAS LIBRERIAS DE SKLEARN

from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
#APLICARLO ES MUY SENCILLO, COGEMOS LA FUNCION, Y LE DECIMOS EL VALOR REAL Y LA PREDICCION
mean_absolute_error(y_test,y_pred) #ERROR BASTANTE GRANDE
```

```python
#PARA HACER EL SQUARE ERROR, ES MUY SIMILAR, PERO CON SQUARE, SALDRA NUM MUCHO MAYOR PORQUE ES TODO AL CUADRADO
mean_squared_error(y_test,y_pred)
```

```python
##TODO ESTO ES TENIENDO EN CUENTA QUE TENEMOS UNA COLUMNA, PERO Y SI TENEMOS VARIAS?
#EL PROCESO ES EXATCAMENTE EL MISMO, VAMOS A USAR UN DATASET DIFERENTES

df_bikes=pd.read_csv('bike.csv')
df_bikes.head()
```

```python
df_bikes.describe()
```

```python
df_bikes.drop(['dteday'], axis=1, inplace=True)
```

```python
##AHORA YA TENEMOS TODOS LOS VALORES NUMERICOS. HAGAMOS LA REGRESION. PRIMERO HACEMOS EL TRAIN TEST SPLIT
#PRIMERO HAY QUE DEFINIR QUE SON LAS X Y QUE SON Y. EN ESTE CASO PREDIGAMOS CNT EN FUNCION DEL RESTO
#VAMOS A COGER TODAS LAS COLUMNAS

x=df_bikes[['instant', 'season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday',
       'weathersit', 'hum', 'windspeed', 'casual',
       'registered', 'cnt']] ##IMPORTANTE EL DOBLE CORCHETE!!!

y=df_bikes['temp'] #VARIABLE QUE QUEREMOS PREDECIR
```

```python
#UNA MANERA MUY FACIL DE ESCOGER VARIABLES X ES CON COLUMNS, COPIANDO Y PEGANDO
df_bikes.columns
```

```python
##SEGMENTAMOS AHORA, SIEMPRE XTRAIN, XTEST, YTRAIN, YTEST!!!

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42) #vamos a usar 80% ahora

print(x_train.shape)
print(y_train.shape) #COMO SOLO ES UN CONJUNTO DE DATOS, LA SOLUCION, NO HAY NINGUNA COLUMNA
print(x_test.shape)
print(y_test.shape)

#VEMOS QUE COINCIDEN 584 Y 584 Y 146 Y 146. 14 COLUMNAS DE 14 VARIABLES
```

```python
## HEMOS DESCOMPUESTO TODO YA, PODRIAMOS HACER UN STANDARSCALER, PERO PRIMERO VAMOS A CONSIDERAR UQ ENO HAY
##HAY QUE CALCULA EL MODELO, LA REGRESION LINEAL EN ESTE CASO

from sklearn.linear_model import LinearRegression
model2=LinearRegression()
model2
```

```python
#AHORA TENEMOS YA LA CAJA NARANJA, AHORA ADAPTAMOS A NUESTROS DATOS (MOMENTO DE ENTRENAMIENTO)

model2.fit(x_train,y_train)
```

```python
##YA ESTA TODO ENTRENADO, AHORA PODEMOS CALCYULAR LOS COEFICIENTES
model2.intercept_ ##PRIMERO INTERCEPT, EL VALOR QUE SIEMPRE VA SOLO
```

```python
## POR OTRO LADO LOS COEFS, PARA CADA UNA DE LAS VARIABLES HAY QUE TENER UN COEFICIENTE
model2.coef_

##AQUI VEMOS VALORES CON NOTACION CIENTIFICA
```

```python
## PARA FACILITAR TODO UN POCO,VAMOS A CREAR UNA LISTA QUE SEA DE REDONEAR A DOS DECIMALES CADA UNO DE LOS COEFICIENTES

coeficientes=[round(x, 2) for x in model2.coef_] #el for es para cada uno de los coeficientes, ESTO ES UN LIST_COMPREHENSSION
coeficientes

#VEMOS QUE TODAS LAS QUE NOS DA 0 SON NO SIGNIFICATIVAS PARA EL MODELO
#PERO SI CAMBIAMOS VARIABLE Y A "TEMP" CAMBIA TODO
```

```python
##PREDECIMOS LA TEMPERATURA AHORA CON PREDICT. VARIABLE NUEVA (PREDICCION) = CON EL DATASET DE TEST

y_pred=model2.predict(x_test)
y_pred
```
