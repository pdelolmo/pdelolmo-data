#Ejercicio: Numero de juegos pares

```python
def es_par(numero):
  if numero%2==0: #el resto, lo que sobra de la vision, si es 0 es par
    result='par'
    return result
  else:
    result='impar'
    return result
```

```python
es_par(8)
```

```python
while True:
  num_usuario=int(input('Introduce un numero '))
  result=es_par(num_usuario)
  if result=='impar':
    print(f'El numero {num_usuario} es impar')
    break
  else:
    print(f'El numero {num_usuario} es par. Intentelo de nuevo')
```

#JOINS

##hay distintas formas de hacer una union. la mas facil es que todos los elementes de una tabla tengan su equivalente en otra tabla, una relacion de 1 a 1. En este caso usamos un INNER

## cuando hablamos del left, coge todos los valores que uno por la izquierda, y si hay elementos que no corresponden con la tabla derecha, los mete pero como nulos

##lo mismo con right

## si no hay relaciones, cojo todo, haya relacion o no se usa OUTTER aunque apenas se usa

```python
df_union=pd.merge(df_clientes, df_ventas, on='ID', how='inner') #en este caso, la relacion es 1 a 1, conviene siempre poner el mismo nombre
df_union.head()
```

```python
def precio_total(row):
  precio_total= row['Cantidad']*row['Precio']
  return precio_total #RECORDAMOS SIEMPRE DEVOLVER EL VALOR

df_union['Precio_total']=df_union.apply(precio_total, axis=1)
df_union.head()
```

```python
def mayor_40(row):
  if row['Edad']>20:
    return 1
  else:
    return 0

df_union['Mayor_40']=df_union.apply(mayor_40, axis=1)
df_union['Mayor_40'].sum()
```

#VISUALIZACION CON PYTHON

```python
#siempre hay que importar 3 librerias matplot, seaborn, plotly

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#estos datos estan en formato JSON, muy utilizado en BBDD como Mongo

from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
diabetes
```

```python
#podemos seleccionar solo la columna 'data'. Es como un diccionario. Es muy importante en este tipo de datasets
#descomponer los datasets, qué definimos como variable a predecir etc

data=diabetes['data']
target=diabetes['target']
feature_names=diabetes['feature_names']
feature_names
```

```python
#VISUALIZAMOS CON MATPLOTLIB, OJO CON NUMPY

import numpy as np

x=data[:,0] #para seleccionar solo la edad de todas las personas
y=target

sorted_indexes=np.argsort(x)
x_sorted=x[sorted_indexes]
y_sorted=y[sorted_indexes] #ahora ya esta todo ordenado

plt.figure(figsize=(10, 6)) #tamaño de figura de matplotlib definida
plt.plot(x_sorted, y_sorted, label='Línea') #dibujamos y vemos el primer grafico
plt.xlabel('Edad normalizada')
plt.ylabel('Pogresión de la Enfermedad')
plt.title('Edad vs Diabetes') #ponemos el titulo
plt.legend() #ponemos la leyenda
plt.grid(True) #ponemos cuadricula debajo
plt.show()



```

```python
#VISUALIZAMOS CON PLOTLY
#SOLO TRABAJA CON DATASETS, HAY QUE TRANSFORMAR LOS DATOS A UN DATASET

df_diabetes=pd.DataFrame({'Edad': x, 'Progresión': y})
df_diabetes.head()

df_diabetes.sort_values('Edad', inplace=True) #ordenamos la variable Edad
df_diabetes.head()

fig=px.line(df_diabetes, x='Edad', y='Progresión', title='Edad vs Diabetes')
fig.show()


```
