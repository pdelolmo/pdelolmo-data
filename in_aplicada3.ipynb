```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_retail=pd.read_excel('online_retail_II.xlsx')
df_retail.head()
df_retail.info()

```

```python
#queremos formatear la fecha usamos pd.to_datetime
df_retail['Date']=pd.to_datetime(df_retail['InvoiceDate'], format='%Y-%m-%d %H:%M:%S') #IMPORTANTE

#hay una forma de que las fechas sean representadas en un caso concreto
#si yo quiero la fecha en formato d-m-Y. tendremos que coger la columna date, aplicarle una funcion, pero como es tan tan pequeña
#no necesitamos un def format x.... cuando son funciones pequeñas se pueden poner con lambdas
df_retail['Date2']= df_retail['Date'].apply(lambda x: x.strftime('%d-%m-%Y'))
df_retail.head()

df_retail['Description'].value_counts() #unique y value_counts para inspeccionar



```

```python
#yo puedo querer que me cuente los atibutos unicos que hay en todo el dtaset, no solo en una columna
#con una funcion. que se llama unique counts que

#CREAMOS UNA FUNCION EN LA QUE PARA CADA COLUMNA DEL DATASET (FOR COLUMNS) VAS A APLICAR EL NUNIQUE AL NOMBRE DE ESA COLUMNA
#Y LOVAS A GUARDAR, DE ESTA FORMA LUEGO HARAS UN PRINT

def unique_counts(df_retail): #le damos a la funcion todo el dataset
  for column in df_retail.columns: #bucle porque quiero coger cada una de las columnas del dataset. para cada una de las columnas en dataset
    count=df_retail[column].nunique() #para saber cuantos valores unicos existen
    print(f'La columna {column} : {count}')

##SIEMPRE QUE ACABAMOS LA FUNCION HAY QUE LLAMARLA!!
unique_counts(df_retail)
```

```python
#vamos a calcular otra columna que va a ser el Precio Total, la cantidad multiplicada por el precio

df_retail['PrecioTotal']=df_retail['Quantity']*df_retail['Price']
df_retail.head()

#ACONSEJABLE SIEMPRE COMPROBEIS EN LA COLUMNA QUE SE CALCULA EL VALOR MAXIMO (PUEDE SER QUE SEA INFINITO)
df_retail['PrecioTotal'].max()

df_retail['Customer ID'].max()

#vamos a filtrar`para ver que personas han comprado (aquellos en los que la cantidad sea > 0), esto me dara V o F
df_retail[df_retail['Quantity'] > 0] #yo solo quiero los verdaderos, sobreescribimos.
df=df_retail[df_retail['Quantity'] > 0] #nombramos un nuevo dataset. ponemos la condicion entre los corchetes del dataset. ya estan los clientes
df.describe() #podemos ver el analisis numerico con describe

#*****

#como calcular la columna de RECENCIA y FRECUENCIA. para ello vanmos a necesitar una nueva libreria, de manejo de tiempos
#datetime

import datetime as dt
NOW=dt.datetime(2025,5,30) #hay formas tambien de coger la fecha actual

#vamos a hacer una tabla y a agrupara por cliente. nos interesa analizar las compras de cada cliente

#queremos calcular una funcion (lambda), la diferencia entre el valor actual y la ultima compra. agg! como apply pero groupby
#pero tambien podemos ver cuant facturas o cuantas compras han realizado.
df.head()

#GROUP BY sera lo mas raro de la clase, agrupar por clientes. hay que calcular las 3 columnas. la de num de dias que pasaron desde la ultima
#compr,a numero de compras que ha realizado y cantidad total de dinero gastada. teniamos que decir la fecha en la que estamos
#NOW=dt.datetime(2025,5,30) pero tambien se puede poner una formula para poner la actual dt.datetime.now(). Queremos agrupar, tenemos una columna
#de customer ID. el gruopby va a fragmentar el dataset en tantas cajas como clientes tenemos. para cada una de las cajas, aplicamos
#una serie de funciones con.agg (a cada una de las cajas vamos a aplicar funciones). la primera es PRecioTotal, cogemos todas las
#si yo quiero calcular cuanto ha gastado cada cliente. tenemos que coger cada una de la columnas de prectiototal y sumarle las
#que corresponden al cliente. de esto ya se encarga el groupby. vamos a coger preciototal y sumarla (x.sum()). con len vemos cuantas filas
#tiene la columna INVOICE (cuantas veces se ha hecho una compra) y dias de cuanto han pasado desde que compro hacemos la resta de dias y lo
#ponemos en formato days (.days)



rfmTable=df.groupby('Customer ID').agg({'Date': lambda x:(NOW-x.max()).days, 'Invoice': lambda x:len(x), 'PrecioTotal': lambda x: x.sum()})
rfmTable

#podemos cambiar el nombre de las columnas. vamos a meter un dict. DATE va a ser RECENCIA y INVOICE FRECUENCIA y PrecioTotal MONETARY

rfmTable.rename(columns={'Date':'Recency', 'Invoice':'Frequency', 'PrecioTotal':'Monetary'}, inplace=True)
rfmTable

#vemos la info
rfmTable.info()




```

```python
#ahora metemos el kmeans

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

rfmTable.head()

#standard scaler nos va a reducir la cantidad de numeros, cuando hay una cantidad muy grande de valores numericos, se reduce
#para que le diferencia sea mejor, aplicamos Scaler para ello, se suele hacer SIEMPRE

#REPASO: standarscaler. primero hablamos de datos (Recencia) que son miles (5k) con datos que son decenas 33, hasta valores muy variados (miles de euros)
#para evitar que el algortimo diga voy a dar mas importancia a recencia (porque son num mas grandes) que a frecuencia (que es mas pequeño)
#hay que estandarizar los datos. a las variables numericas las etsandarizamos. con un proceso que es STANDARSCALER
#se importan las librerias, y importamos el standarcaler y ajustamos a los datos que tenemos y los transformamos para no hablar de 5k y 33
#sino que esta todo en la misma escala (Se puede comparar mejor). sirve para que el algortimo de la misma importancia a todas las variables

scaler=StandardScaler() #cunado le demos datoa a este standar scaler empezara a estar en azul
scaler
```

```python
rfmTable.head()
```

```python
x_scaled=scaler.fit_transform(rfmTable)
x_scaled
```

```python
#vamos a predefinir que sean 3 grupos, aunque se puede hacer con metodo del codo

cluster=KMeans(n_clusters=3) #parecido al standarscaler pero hay que indicar cuantos quieres. lo vamos a llamar cluster

#pero tenemos que pasarle los datos, los datos son valores numericas transformadas con el fit
cluster.fit(x_scaled) # sin x_scaled sale amarillo, con el fit se hace una agrupacion con los valores que se ha dado (azul). ya ajustado

#si aplicamos cluster.labels ya me ha agrupado los puntos en 3.

rfmTable['cluster_new']=cluster.labels_     # creamos una nueva columna y que sean los grupos que me ha dado (cluster.labels)
rfmTable

rfmTable['cluster_new'].value_counts() #contamos con value_counts

#alrogitmo de ML. me ha clasificado 5 elementos en grupo 2, 3200 en el grupo 0 que tienen semejanzas entre si y 1000 en el grupo 1
#ANTES DE APLICARLO
```
